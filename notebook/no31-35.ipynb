# No30

```python
def neko_lines():

    with open('./neko.txt.mecab') as f:
        morphemes = []
        for line in f:
            cols = line.split('\t')
            if (len(cols) < 2):
                raise StopIteration
            res_cols = cols[1].split(',')

            morpheme = {
                'surface': cols[0],
                'base': res_cols[6],
                'pos': res_cols[0],
                'pos1': res_cols[1],
            }

            morphemes.append(morpheme)
            
            # 句点で一文と判定させる
            if res_cols[1] == '句点':
                # yieldで一文をシーケンスとして持たせてその後変数を初期化する
                yield morphemes
                morphemes = []
lines = neko_lines()
for i, line in enumerate(lines):
#     print(line)
    if i > 50:
        break
```

```python
def firstn(n):
    num = 0
    while num < n:
        yield num
        num += 1
for i in firstn(10):
    print(i)
```
